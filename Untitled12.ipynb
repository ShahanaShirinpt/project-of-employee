```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/employee_dataset.csv.csv")
```

```python
df.head(20)
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Check for missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
# Drop rows with any null values
df.dropna(axis=0, inplace=True)
```

```python
# Drop columns with any null values
df.dropna(axis=1, inplace=True)
```

```python
# Check if any null values are left in rows
print(df.isnull().any(axis=1).sum())
```

```python
# Check if any null values are left in columns
print(df.isnull().any(axis=0).sum())
```

```python
df.head(458)
```

```python
print("\nMissing Values After Preprocessing:")
print(df.isnull().sum())
```

```python
# Correct the 'height' column
df['height'] = np.random.uniform(150, 180, size=len(df))
```

```python
# Find the number of unique teams
num_teams = df['Team'].nunique()

print(f'Total Number of Teams: {num_teams}')
```

```python
# Task 1: Number of employees in each team and percentage splitting
team_stats = df['Team'].value_counts()
percentage_split = team_stats / len(df) * 100
team_info = pd.DataFrame({'Count': team_stats, 'Percentage': percentage_split})
print("\nTask 1: Employees in Each Team and Percentage Splitting:")
print(team_info)
```

```python
# Task 2: Segregate employees by position
position_stats = df['Position'].value_counts()
print("\nTask 2: Employees Segregated by Position:")
print(position_stats)
```

```python
# Task 3: Find the age group with the most employees
age_groups = pd.cut(df['Age'], bins=[20, 30, 40, 50, 60, 70])
age_group_stats = age_groups.value_counts()
print("\nTask 3: Age Group with Most Employees:")
print(age_group_stats)
```

```python
# Task 4: Identify team and position with high spending in terms of salary
high_spending = df.groupby(['Team', 'Position'])['Salary'].sum().idxmax()
print("\nTask 4: Team and Position with High Spending:")
print(high_spending)
```

```python
correlation = df['Age'].corr(df['Salary'])
print(f"\nTask 5: Correlation Between Age and Salary: {correlation:.2f}")

# Visualize the correlation
sns.scatterplot(x='Age', y='Salary', data=df)
plt.title(f'Correlation: {correlation:.2f}')
plt.show()
```
